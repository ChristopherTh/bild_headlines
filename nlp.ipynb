```python
import nltk
import pandas as pd
from collections import Counter
import pandas as pd
import seaborn as sns
sns.set(rc={'figure.figsize':(20,12)})
import matplotlib.pyplot as plt
import string
from nltk.corpus import stopwords
```

```python
df = pd.read_csv("headlines_cleaned.csv")
text = df.text.str.replace(" ", "", 1).to_list()
```

```python
satzzeichen = list(string.punctuation)
satzzeichen.remove('!')
satzzeichen.remove('?')

german_stop_words = stopwords.words('german')

ignore_list = german_stop_words + satzzeichen
```

```python
tokenized_sent = [nltk.tokenize.word_tokenize(i,language='german') for i in text]
```

```python
cnt = Counter()
for sublist in tokenized_sent:

    for i in sublist:

        if i not in ignore_list:

            cnt[i.lower()] += 1
```

```python
df_common = pd.DataFrame(cnt.most_common(1000), columns = ['Wort', 'Anzahl'])
df_common.set_index("Wort", inplace = True)
```

```python
df_common.iloc[:50].plot(kind = 'bar')
```
