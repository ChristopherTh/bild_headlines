```python
import pandas as pd
```

```python
df = pd.read_csv("raw_headlines.csv")
df
```

```python
# Ich habe beim Datenzug angenommen das alle Monate 31 Tage haben
# falls ein Monat weniger als 31 Tage hatte wurde man auf eine Fehlerseite verwiesen
# in dem Fall aheb ich ein kennzeichen gesetzt welches ich hier rausfiltere
df1 = df.loc[df.link != 'exeption']
df1
```

```python
df1_link = df1.link.str.split('/', expand = True)
df1_link
```

```python
df1_link.drop(labels = 0, axis = 1, inplace = True) # leere Spalte
df1_link
```

```python
df2 = pd.concat([df1, df1_link], axis = 1).drop(columns = 'link')
df2
```

```python
# die nächsten SChritte sind dazu da alle Zeiteinheiten in einer pandas datetime Spalte zusammenzufassen
df2['hour'] = df2['Unnamed: 3'].str.slice(start = 0, stop = 2)
df2['minute'] = df2['Unnamed: 3'].str.slice(start = 3, stop = 5)
df2
```

```python
df2.rename(columns={'Unnamed: 0': 'year', 'Unnamed: 1': 'month', 'Unnamed: 2': 'day'}, inplace=True)
```

```python
df2.index = pd.to_datetime(df2[['year', 'month', 'day', 'hour', 'minute']])
df2.drop(columns = ['Unnamed: 3', 'year', 'month', 'day', 'hour', 'minute'],inplace = True)
df2.rename(columns = {1:'link1', 2:'link2', 3: 'link3', 4:'link4',5:'link5'},inplace = True)
df2
```

```python
df2.reset_index().to_csv('headlines_cleaned.csv')
```
